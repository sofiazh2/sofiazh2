### 1. Encabezados

Se usa el símbolo `#` para crear encabezados. Cuantos más `#`, más bajo el nivel del encabezado.

# Encabezado 1

## Encabezado 2

### Encabezado 3

#### Encabezado 4

## <span style="color:LightBlue">2. Personalización del texto</span>

### 2.1.  Texto en negrita

Se usa el símbolo `**` o `__` para negrita.

**Texto en negrita**
__Texto en negrita__

### 2.2. Texto en cursiva

Se usa el símbolo `_` o `*` para cursiva. 

_Texto en cursiva_
*Texto en cursiva*

### 2.3. Combinar negrita y cursiva

Se pueden combinar negrita y cursiva usando tres asteriscos `***` o guiones bajos `___`.

***Texto en negrita y cursiva***
___Texto en negrita y cursiva___

### 2.4. Tachar texto

Para tachar un texto, se usa `~~`.

~~Este texto está tachado~~

### 2.5. Colorear texto (usando HTML)

Resulta, sucede y acontece que Markdown no soporta colores por defecto, pero se puede incluir HTML para cambiar el color del texto, así:

<span style="color:red">Este texto es rojo</span>

### 2.5. Subíndices y superíndices (HTML)

Si en algún momento necesitamos escribir subíndices o superíndices, podemos usar HTML:

Agua: H<sub>2</sub>O  
Cuadrado: x<sup>2</sup>

### 2.6. Escapar caracteres especiales

Si necesitamos mostrar caracteres que tienen un significado especial en Markdown (como `#`, `*`, `_`, etc.), se usa una barra invertida `\` para "escaparlos".

\*Este texto no será en cursiva\*
\# Este texto no será un encabezado

### 2.7. Ecuaciones en LaTeX

Se usa `$` para escribir ecuaciones en línea y `$$` para ecuaciones en bloques.

Esto es una ecuación en línea:

$E=mc^2$

Esto es una ecuación en bloque:

$$ E=mc^2 $$

## <span style="color:LightBlue">3. Listas</span>

Para listas ordenadas se usan números seguidos de un punto. Por otro lado, para listas desordenadas se usan `-`, `*` o `+`.

1. Primer ítem
2. Segundo ítem
   1. Sub-ítem

- Primer ítem
* Segundo ítem

### 3.1. Anidación de listas

Podemos anidar listas ordenadas y desordenadas dentro de otras listas añadiendo espacios adicionales.

1. Primer ítem
   - Sub-ítem 1
   - Sub-ítem 2
2. Segundo ítem
   * Sub-ítem 3

### 3.2. Listas de contenido automáticas

En documentos largos, es útil tener una tabla de contenido automática. Algunas extensiones como **Markdown All in One** permiten crear una con solo escribir `[[toc]]`.

[[toc]]

Esto generará una tabla de contenido basada en los encabezados.

### 3.3. Ítems de tareas

Se pueden crear listas de tareas utilizando corchetes `[ ]` para ítems pendientes y `[x]` para ítems completados.

- [ ] Tarea pendiente
- [x] Tarea completada

## <span style="color:LightBlue">6. Tablas y diagramas</span>

### 6.1. Tablas

Se usa `|` para definir las columnas y `-` para separar el encabezado del contenido.

| Columna 1 | Columna 2 |
| --------- | --------- |
| Valor 1   | Valor 2   |

### 6.2. Diagramas con Mermaid.js

Si queremos diagramas de flujo o gráficos, podemos usar **Mermaid.js** dentro de Markdown, especialmente si tenemos instaladas extensiones como Markdown Preview Enhanced. Aquí un ejemplo de un diagrama de flujo:

```mermaid
graph TD;
    A-->B;
    A-->C;
    B-->D;
    C-->D;
```

## <span style="color:LightBlue">8. Líneas horizontales y saltos de línea</span>

### 8.1. Líneas horizontales

Se usan tres o más guiones `---`, asteriscos `***`, o guiones bajos `___`.

---
***
___

### 8.2. Saltos de línea

Para insertar un salto de línea, dejamos dos espacios al final de la línea antes de presionar "Enter". Si no añadimos estos espacios, Markdown unirá las líneas.

Esto es una línea con un  
salto de línea.

## <span style="color:LightBlue">10. Extensiones de Visual Studio Code</span>

Aparte de la extensión de Jupyter, hay otras que pueden mejorar tu flujo de trabajo:

- **Markdown All in One**: Proporciona atajos de teclado, tablas de contenido automáticas, y otras herramientas útiles.

- **Markdown Preview Enhanced**: Añade vistas previas mejoradas, diagramas interactivos y exportación a diferentes formatos (PDF, HTML, etc.).

- **Markdown Lint**: Te ayuda a mantener un estilo coherente y te alerta sobre errores comunes.

## <span style="color:LightBlue">11. Atajos de teclado útiles en Visual Studio Code</span>

- **Ctrl + K + V**: Abrir una vista previa del archivo Markdown.
