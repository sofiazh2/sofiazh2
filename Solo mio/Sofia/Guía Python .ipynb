## <span style="color:LightSalmon">5. Buenas prácticas para escribir código en Python</span>

### 5.1. Variables

Las variables se nombran usando la convención *snake_case*.
  
**Reglas:**
  
  - Usa letras minúsculas separadas por guiones bajos (`_`).
  
  - El nombre debe describir claramente el propósito de la variable.
  
  - Evita nombres de una sola letra a menos que sea en bucles (`i`, `j`, etc.).



```python
# Ejemplo:
edad_usuario = 25
nombre_completo = "Juan Pérez"

# Mala práctica:
EdadUsuario = 25
nc = "Juan Pérez"
```

### 5.2. Constantes

Las constantes se nombran en UPPER_CASE.

**Reglas:**
  
- Usa letras mayúsculas separadas por guiones bajos (_).
  
- Las constantes son valores que no deberían cambiar durante la ejecución del programa.


```python
# Ejemplo:
PI = 3.1416
MAX_INTENTOS = 5

# Mala práctica:
pi = 3.1416
MaxIntentos = 5
```

### 5.2. Funciones

Las funciones también se nombran en snake_case.

**Reglas:**

- Usa nombres en minúsculas con guiones bajos (_) para separar las palabras.
  
- El nombre debe describir claramente lo que hace la función.
  
- Usa verbos que indiquen la acción que realiza la función.


```python
# Ejemplo de función pública:
def calcular_suma(a, b):
    return a + b

# Mala práctica:
def CalcularSuma(a, b):
    return a + b
```

### 5.3. Métodos (en clases)
   
Los métodos se nombran igual que las funciones, usando snake_case.

**Reglas:**

- Usa nombres descriptivos que reflejen lo que hace el método.

- Los métodos privados o internos deben comenzar con un guion bajo (_).

Ejemplo:

```python
# Ejemplo de método privado:
class Vehiculo:
    def _verificar_nivel_combustible(self):
        print("Nivel de combustible verificado")
```

Ejemplo de lo que no se debe hacer:

```python
class Coche:
    def IniciarMotor(self):  # No uses camelCase.
        print("Motor iniciado")

    def ComprobarBateria(self):  # Usa un guion bajo para métodos internos.
        print("Batería OK")
```

### 5.4. Notación de clases

Las clases se nombran en PascalCase (también llamada CapWords o CamelCase).

**Reglas:**

- Cada palabra comienza con una letra mayúscula, y no se utilizan guiones bajos.

- Los nombres deben ser descriptivos.
  

```python
# Ejemplo:
class CocheElectrico:
    pass

# Mala práctica:
class coche_electrico:
    pass
```
